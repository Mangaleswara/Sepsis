# Outlier removal

Removing values which are greater than 3 times standard deviation

```python
def replace(group, stds):
    group[np.abs(group - group.mean()) > stds * group.std()] = np.nan
    return group
```

```python
df1=df.loc[:, df.columns] = df.transform(lambda g: replace(g, 3))
```

```python
df1.isnull().sum()
```

# Formula Imputation

```python
df1["MAP"].fillna((0.33*(df1["SBP"]-df1["DBP"]))+df1["DBP"],inplace=True)
df1["SBP"].fillna((3*(df1["MAP"]-(0.66*(df1["DBP"])))),inplace=True)
df1["DBP"].fillna((1.5*(df1["MAP"]-(0.33*(df1["SBP"])))),inplace=True)
```
